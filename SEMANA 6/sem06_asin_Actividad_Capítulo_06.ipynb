```python
#Explique brevemente y en forma clara qué es un algoritmo Greedy y como funciona
"""
Un algoritmo Greedy es un enfoque de resolución de problemas que sigue el principio de 
tomar la mejor opción local en cada paso" con la esperanza de que esta estrategia conduzca a
una solución óptima global. En otras palabras, en cada paso, el algoritmo 
toma la decisión óptima en ese momento sin considerar las consecuencias futuras.
Funciona seleccionando la mejor opción disponible en cada paso hasta que se alcanza una solución.
"""

#Enumera algunas ventajas.
"""
1.Es simple y fácil de implementar.
2.Tiende a ser eficiente en términos de tiempo de ejecución.
3.Puede ser útil para encontrar soluciones aproximadas en problemas de optimización.
"""

#Enumera algunas desventajas.
"""
1.No garantiza una solución óptima en todos los casos.
2.Puede llevar a soluciones subóptimas si la elección local no conduce a la mejor solución global.
3.Requiere una verificación adicional para asegurar que la solución encontrada sea válida y óptima.
"""

#Escriba el código desarrollado con el enfoque Greeddy para el problema de cambio de monedas, haga diferentes pruebas cambian las entradas (recuerde que las entrada de este algoritmo son las monedas disponibles y la cantida que se desea cubrir, con estas entradas el algoritmo debe retornar la combinación de monedas que cubre el monto solicitado, minimizando la cantidad de monedas usadas)
def cambio_monedas(monedas_disponibles, cantidad):
    monedas_usadas = []
    monedas_disponibles.sort(reverse=True)  # Ordenar las monedas de mayor a menor valor
    for moneda in monedas_disponibles:
        while cantidad >= moneda:
            cantidad -= moneda
            monedas_usadas.append(moneda)
    return monedas_usadas


#Ahora pruebe los resultados con siguientes entradas: monedas_disponibles = [11, 5, 1] para cubrer la cantidad de 15, observe los resultados, haga algunas analísis de lo que pasaria con esta combinación de entradas.
monedas_disponibles = [11, 5, 1]
cantidad = 15
resultado = cambio_monedas(monedas_disponibles, cantidad)
print("Combinación de monedas para cubrir", cantidad, ":", resultado)








```

## Instrucciones


1. Enumera las diferencias entre los dos efoques.
Enfoque de solución: La programación dinámica resuelve problemas dividiéndolos en subproblemas superpuestos y construyendo soluciones a partir de estos subproblemas. Divide y vencerás resuelve problemas dividiéndolos en subproblemas no superpuestos y combinando sus soluciones.
Uso de memoria: La programación dinámica utiliza más memoria que Divide y Vencerás porque almacena las soluciones de todos los subproblemas para su reutilización.
Tipo de problema: La programación dinámica es útil cuando los subproblemas se superponen, mientras que Divide y Vencerás es más adecuado para problemas en los que los subproblemas son independientes.
2. Nombre algunos casos donde funcionam mejor la programación dinámica.
