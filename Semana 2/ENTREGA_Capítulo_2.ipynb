De la unidad 10, sección 6.3 revise los métodos disponibles para usar sobre las listas.

Aplique los métodos disponibles sobre la lista mostrada en la siguiente celda. 

Nota: usando el método **sort()** ordene de mayor a menor y viceversa (Si es necesario revise documentación adicional relacionada sobre este método)


````python

list_notas = [15, 21, 24, 40, 13, 14, 18, 13, 17, 14.5]

# Método append() - Agregar un elemento al final de la lista
list_notas.append(22)
print("Después de append(22):", list_notas)

# Método insert() - Insertar un elemento en una posición específica
list_notas.insert(3, 30)
print("Después de insert(3, 30):", list_notas)

# Método remove() - Eliminar la primera ocurrencia de un elemento específico
list_notas.remove(13)
print("Después de remove(13):", list_notas)

# Método pop() - Eliminar y devuelve el elemento en una posición específica
elemento_eliminado = list_notas.pop(5)
print(f"Después de pop(5) - Elemento eliminado: {elemento_eliminado}, Lista resultante:", list_notas)

# Método index() - Devolver el índice de la primera ocurrencia de un elemento
indice_17 = list_notas.index(17)
print(f"Índice de la primera ocurrencia de 17: {indice_17}")

# Método count() - Devolver el número de ocurrencias de un elemento
ocurrencias_14 = list_notas.count(14)
print(f"Número de ocurrencias de 14: {ocurrencias_14}")

# Método sort() - Ordenar la lista de menor a mayor
list_notas.sort()
print("Después de sort() (menor a mayor):", list_notas)

# Método reverse() - Invertir  el orden de la lista
list_notas.reverse()
print("Después de reverse():", list_notas)













````

# Actividad: Crear diccionarios y guardarlos como archivo JSON.

De la unidad 11 sección 4.6 revise el ejemplo mostrado.

Genera un diccionario, sobre la información que deseas guardar como base de datos, desarrollo el código necesario para generar el archivo JSON, y también el código para cargar la información del archivo JSON.


````python
import json

# Crear un diccionario con información de una base de datos
base_de_datos = {
    'usuarios': [
        {
            'id': 1,
            'nombre': 'Usuario1',
            'correo': 'usuario1@example.com'
        },
        {
            'id': 2,
            'nombre': 'Usuario2',
            'correo': 'usuario2@example.com'
        }
    
    ]
}

# Guardar el diccionario como un archivo JSON
nombre_archivo = 'base_de_datos.json'
with open(nombre_archivo, 'w') as archivo:
    json.dump(base_de_datos, archivo, indent=2)  # 'indent' es opcional, pero mejora la legibilidad del archivo JSON

print(f"Archivo JSON '{nombre_archivo}' creado exitosamente.")

# Cargar la información desde el archivo JSON
with open(nombre_archivo, 'r') as archivo:
    datos_cargados = json.load(archivo)

# Mostrar la información cargada
print("\nInformación cargada desde el archivo JSON:")
print(json.dumps(datos_cargados, indent=2))  # Mostrar los datos cargados en formato JSON














````

# Actividad: Revisión y manejo de listas bidimensionales.

## Instrucciones

1. Describa brevemente que entiende por lista bidimensional.
2. Crear una lista bidimensional.
3. Generar un doble loop para acceder de la lista bidimensional
4. Sumar todo los elementos de esta lista bidimensional.

````python
# 1.Describa brevemente que entiende por lista bidimensional.

#Lista Bidimensional:
#Una lista bidimensional, también conocida como matriz, 
#es una estructura de datos en la que los elementos están organizados en filas y columnas. 
#Puedes pensar en ello como una lista de listas, donde cada lista interna 
#representa una fila y los elementos dentro de esa lista son los elementos de la fila.

# 2. Crear una lista bidimensional
lista_bidimensional = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

# 3. Generar un doble bucle para acceder a los elementos
filas = len(lista_bidimensional)
columnas = len(lista_bidimensional[0])

# 4. Imprimir la lista bidimensional
print("Lista Bidimensional:")
for i in range(filas):
    for j in range(columnas):
        print(lista_bidimensional[i][j], end=" ")
    print()

# 5. Sumar todos los elementos de la lista bidimensional
suma_total = 0
for i in range(filas):
    for j in range(columnas):
        suma_total += lista_bidimensional[i][j]

print("Suma total de los elementos:", suma_total)






````
