## Q 04-01

### ¿Qué no es correcto como interpretación del modelo de ajuste en el siguiente análisis estadístico?



1. En el análisis estadístico, la prueba unilateral indica que la hipótesis nula para el parámetro es H_0 ∶ θ = θ_0 y la hipótesis alternativa se puede definir como H_1 ∶ θ < θ_0 o H_1 ∶ θ > θ_0.
2. El error tipo I es el nivel de significancia y es un error que ocurre cuando se rechaza la hipótesis nula verdadera aunque la hipótesis nula sea verdadera.
3. Si la probabilidad de significancia p es mayor que el nivel de significancia α (p > α), se rechaza la hipótesis nula y se adopta la hipótesis alternativa
4. Para probar la significancia de la ecuación de regresión, se debe realizar una prueba de significancia en la pendiente de la ecuación de regresión, por lo que la hipótesis es H_0 ∶ β_1= 0, H_1 ∶ β_1≠ 0 (donde β_1 es la pendiente de la ecuación de regresión ecuación).

**Respuesta**

Si la probabilidad de significancia p es mayor que el nivel de significancia α (p > α), se rechaza la hipótesis nula y se adopta la hipótesis alternativa.

En realidad, si la probabilidad de significancia p es menor que el nivel de significancia α (p < α), 
entonces se rechaza la hipótesis nula y se adopta la hipótesis alternativa. Si p > α, no se puede rechazar la hipótesis nula.

## Q 05-02

### Suponiendo que me caso y tengo tres hijos, suponiendo que la variable aleatoria X es el número de hijos, ¿cuál es la probabilidad, la media y la varianza de tener dos hijos en este caso? (En este caso, se supone que la probabilidad de tener un hijo y una hija es la misma).

**Respuesta**

# Probabilidad de tener dos hijos varones:
La distribución binomial es adecuada para este caso, donde n = 3 (número de ensayos) y p = 0.5 (probabilidad de éxito, es decir, tener un hijo varón).
La probabilidad de tener exactamente dos hijos varones se calcula utilizando la fórmula de la distribución binomial: P(X = 2) = (3 eligiendo 2) × (0.5)^2 × (0.5)^1 P(X = 2) = (3 × 2 × 1) / (1 × 2 × 3) × (0.5)^2 × (0.5)^1 P(X = 2) = 3/8 ≈ 0.375
## Media de la distribución:
La media de una distribución binomial se calcula como μ = n × p
En este caso, n = 3 (número de ensayos) y p = 0.5 (probabilidad de éxito)
μ = 3 × 0.5 = 1.5
## Varianza de la distribución:
La varianza de una distribución binomial se calcula como σ^2 = n × p × (1 - p)
en este caso, n = 3 y p = 0.5
σ^2 = 3 × 0.5 × (1 - 0.5) = 0.75
# En resumen:

Probabilidad de tener dos hijos varones: 0.375 (o 3/8)
- Media de la distribución: 1.5
- Varianza de la distribución: 0.75

## Q 05-05

### Entre los estudiantes que tomaron la clase de Información y Estadística, se seleccionaron aleatoriamente 30 estudiantes. Al investigar el promedio de sus calificaciones, la media fue 80 y la varianza fue 9. Encuentre un intervalo de confianza del 95% para el promedio de sus calificaciones.

Intervalo de confianza = media muestral ± (valor crítico) * (error estándar)

Donde:

Media muestral = 80 (dado)
Valor crítico = valor correspondiente al nivel de confianza deseado en la distribución normal estándar
Error estándar = desviación estándar de la muestra / raíz cuadrada del tamaño de la muestra
Valor crítico = 1.96 (para un nivel de confianza del 95%)
Desviación estándar de la muestra = raíz cuadrada de la varianza = raíz cuadrada de 9 = 3
Error estándar = 3 / raíz cuadrada de 30 = 3 / 5.48 = 0.547

Intervalo de confianza = 80 ± (1.96) * (0.547)
Intervalo de confianza = 80 ± 1.072
Intervalo de confianza = (78.928, 81.072)

Por lo tanto, el intervalo de confianza del 95% para el promedio de las calificaciones de los estudiantes es (78.928, 81.072).

Esto significa que podemos estar 95% seguros de que el promedio de las calificaciones de todos los estudiantes que tomaron la clase de Información y Estadística se encuentra entre 78.928 y 81.072.
